### Gráfico de barras

```python
plt.bar(range(1,11),
        [2,4.5,1,2,3.5,2,1,2,3,2],
        label = "Jim",
        color = "m", # m para magenta
        align = "center")

plt.title("Results")
plt.xlabel("Semester")
plt.ylabel("Grade")

plt.legend()
plt.grid(True, color="y")
```

### Agregando otra barra en el gráfico

### Etiquetar nombres en los valores

```python
rainfall = [17,9,16,3,21,7,8,4,6,21,4,1]

months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']

plt.bar(months, rainfall, align = 'center', color = 'orange')
plt.show()

# No se ordena alfabeticamente porque es un bug en matplotlib < 2.2.0 y yo tengo 2.2.4
```

```python
rainfall = [17,9,16,3,21,7,8,4,6,21,4,1]
months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']

plt.bar(range(len(rainfall)), rainfall, align = 'center', color = 'orange')
plt.xticks(range(len(rainfall)), months, rotation = 'vertical')
plt.show()
```

## Diagrama de pastel

Gráfico estadístico en forma circular muy usado en la representación de datos para ilustrar proporciones.

```python
labels = ["Chrome", "Internet Explorer", "Firefox","Edge","Safari","Sogou Explorer","Opera","Others"] # buscadores

marketshare = [61.64, 11.98, 11.02, 4.23, 3.79, 1.63, 1.52, 4.19] # porcentajes de personas.

explode = (0,0,0,0,0,0,0,0)

plt.pie(marketshare,
        explode = explode,  # que tan lejos del centro empezaremos cada cuña
        
        labels = labels, # etiqueta 
        autopct="%.1f%%", # tamaño de porcentaje
            
        shadow = True,
        startangle = 45) # movernos en forma antihorario

plt.axis("equal")
plt.title("Web browser marketing - 2018")
plt.show()
```

### Retirando una parte del gráfico 

Vamos a usar el parámetro explode para lograr esto.

```python
labels = ["Chrome", "Internet Explorer", "Firefox","Edge","Safari","Sogou Explorer","Opera","Others"] # buscadores

marketshare = [61.64, 11.98, 11.02, 4.23, 3.79, 1.63, 1.52, 4.19] # porcentajes de personas.

explode = (0,0,0.5,0,0.8,0,0,0)

plt.pie(marketshare,
        explode = explode,  # fraccion del ratio con el cual compezaremos cada cuña
        
        labels = labels, # etiqueta 
        autopct="%.1f%%", # tamaño de porcentaje
            
        shadow = True,
        startangle = 45) # movernos en forma antihorario

plt.axis("equal")
plt.title("Web browser marketing - 2018")
plt.show()
```
