```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('olympics.csv')
df.head()
```

#### Quitamos la primera fila 

```python
df = df[1:]
df.head()
```

#### Le agregaremos nombres a las columnas

```python
index = np.array(['Country','Summer','Gold','Silver','Bronze','Total','Winter','Gold','Silver','Bronze','Total','Games','Gold','Silver','Bronze','Combined Total'])
```

```python
df.columns = index
df.head()
```

### Ejercicio : Separamos Country como Country name y Country code

Aquí primero separamos el nombre con un codigo (el code aún no esta del todo limpio)

```python
df['Country name'], df['Country code'] = df['Country'].str.split("(",1).str
df.head(10)
```

Terminamos del limpiar el codigo y lo que queda podemos mandarlo a una columna trash.

```python
df['Country code'], df['Trash'] = df['Country code'].str.split(")",1).str
df.head(10)
```

### Otra manera

Solo para sacar country names.

```python
country_names = [x.split('\xc2\xa0(')[0] for x in df.iloc[:, 0]]
df['Country name'] = country_names
df.head()
```

Borramos el trash para quedarnos con los datos que sirven.

```python
df = df.drop('Trash', axis = 1)
df.head()
```
