## Analizando el marco de dato

Otra buena pr√°ctica es analizar obviamente nuestro dataset, que ya se realizo en el ejercicio 1:


* mpg (continous): Miles Per Gallon o en espa√±ol millas por galeon es decir el recorrido por galeon (en millas) de nuestro carro.


* cylinders (multi-valued discrete): la cantidad de cilindros que tiene nuestro carro. En general, un motor con m√°s cilindros produce m√°s potencia, mientras que un motor con menos cilindros obtiene una mejor econom√≠a de combustible.


* displacement (continuous): El desplazamiento del motor es el volumen barrido de todos los pistones dentro de los cilindros de un motor alternativo en un solo movimiento desde el punto muerto superior (TDC) al punto muerto inferior (BDC). Se especifica com√∫nmente en pulgadas c√∫bicas.


* hoursepower (continuous): la potencia del motor del carro en caballos de fuerza


* weight (continuous): peso del carro (al parecer en kg)


* acceleration (continuous): la aceleraci√≥n de nuestro coche, tambi√©n podemos decir que es el tiempo en segundos que nuestro coche alcanza una velocidad de 60mph desde el reposo.


* year (multi-valued discrete): a√±o de fabricaci√≥n, al parecer de los a√±os 1970 hacia adelante.


* origin (multi-valued discrete): Aqui discerniendo por el modelo del carro (gracias a la columna name) podemos decir que..

    * 1: USA
    * 2: Europa
    * 3: Japon


* name (string):  nombre del carro

### Creando variable binaria mpg01

Piden exactamente hallar si un carro tiene m√°s o menos gas mileage que ser√≠a traducido como rendimiento de gasolina, esto es agarrar mpg como se hizo en el ejercicio uno por dos razones:

* Ninguna otra columna se relaciona con el rendimiento de gas directamente como "millas por galeon"
* A m√°s millas por galeon m√°s rendmiento de nuestro gasolina.

A partir de la media de la columna mpg vamos a filtrar de esta manera:

   * Si es mayor que la media ser√° 1
   * Caso contrario sera 0

```python
mpg_mean = df.mean()[0]
print("mpg: "+str(mpg_mean))

df["mpg01"] = (df["mpg"]>mpg_mean).astype(int)
df.iloc[[12,14]] # ilustramos que si nos salio bien
```

### Preparandonos para el cross-validation

```python
from sklearn.model_selection import train_test_split,cross_val_score
```

### Columna name

Aqui se nos presenta un problema interesante debido a que la columna name de nuestro marco de datos son etiquetas:

```python
df["name"].head()
```

Se dice entonces que nuestra columna name es de clase categ√≥rica por lo que podriamos convertirlo a datos num√©ricos entendibles para nuestro modelo de esta forma:

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df["name_clas"] = labelencoder.fit_transform(df.name.values)
df.head()
```

Y como se puede ver a cada carro le dio un valor por ejemplo "chevrolet chevelle malibu" tiene el numero 48,
pero aqui tenemos otro problema, el modelo puede tomar estos valores de clasificaci√≥n no como una etiqueta num√©rica sino como un valor (por ejemplo: se confunde con hay 48 chevrolet chevelle malibu") o con algun tipo de jerarqu√≠a por lo que una soluci√≥n es usar one-hot encoder.

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder(categories='auto')
x = onehotencoder.fit_transform(df.name.values.reshape(-1,1))
x.toarray().shape # aqui tenemos 304 diferentes valores y los dem√°s son repetitivos por los que no lo gener√≥.
```

Pero al tener 304 columnas en vez de 1 sola hace que nuestro modelo SVM tenga demasiadas variables lo que no es correcto al hacer el ajuste.

**Decisi√≥n:** No tomar esta columna name.

```python
df = df.drop(columns=["name_clas"])
```

### Variable categ√≥rica: origin

El mismo problema que con name tenemos con origin, lo bueno que esta vez solo son 3 origenes por lo que podemos crear variables dummies para poder trabajar con este feature

```python
df_origin = pd.get_dummies(df["origin"],prefix='origin')
df_origin.tail()
```

```python
df = df.drop(columns=["origin"])
df = pd.concat([df,df_origin],axis = 1)
df.head()
```

```python
# Como usaremos ahora mpg01 como clases para ser clasificados nuestros features ya no hay necesidad de mpg 

columns = ['cylinders', 'displacement', 'horsepower', 'weight','acceleration', 'year',
           'origin_1', 'origin_2', 'origin_3']
```

### Realizando CrossValidation 

#### 10-CrossValidation

10-CrossValidation para cada valor de COST (C)

```python
X = df[columns]
Y = df["mpg01"]
```

```python
from sklearn import svm
import matplotlib.pyplot as plt
```

### Caso por defecto c = 1

```python
clf = svm.SVC(kernel='linear',C=1)

scores = cross_val_score(clf,X,Y,cv = 10)
print("mean: "+str(scores.mean()))
print("std: "+str(scores.std()))
```

Ahora podemos ver como haciendo un 10-cross validation pero con un cost = 1 la media del score es 0.881521 aproximadamente... lo que queda ver es como variando este c obtenemos un mejor score ... si es que se puede.

### Variando c

```python
for i in range(2,10):
    clf = svm.SVC(kernel='linear',C=i)
    scores = cross_val_score(clf,X,Y,cv = 10)
    print("C= %d, Media: %0.16f , Std: %0.16f)" % (i, scores.mean(), scores.std()))
```

```python
cc = [2,3,4,5,6,7,8,9]
median = [0.8790856950067477, 0.8765215924426452, 0.8790856950067477, 0.8765856950067477,0.8765856950067477,
          0.8790856950067477, 0.8715856950067478, 0.8740856950067478]

print("max: "+str(max(median)))
plt.plot(cc,median,"r")
```

Si bien cuando c = 2 , 4 y 7 tenemos el mejor score... luego la tendencia es hacia abajo a medida que aumentamos el c

```python
for i in range(15,21):
    clf = svm.SVC(kernel='linear',C=i)
    scores = cross_val_score(clf,X,Y,cv = 10)
    print("C = %d, Media: %0.16f , Std: %0.16f)" % (i, scores.mean(), scores.std()))
```

```python
cc = [15, 16, 17, 18, 19, 20]
median = [ 0.8740856950067478, 0.8740856950067476, 0.8740856950067478, 0.8715856950067478, 0.8740856950067478,
         0.8740856950067478]

print("max: "+str(max(median)))
plt.plot(cc,median,"r",color="blue")
```

Al parecer esta tendencia hacia abajo sigue su curso pues en este sector (15-20) el m√°ximo c es 0.8740 que es menor que el m√°ximo c en la region anterior (que fue 0.87908)

#### Nota:
Aqui vemos como los valores no cambian mucho a pesar de usar un 10-crossvalidation por lo que podemos decir el sobreajuste es bajo y nuestro modelo es optimo, funcion√≥ tomar origin como variables categ√≥ricas !

### Valores c peque√±os < 1

```python
cant = 11

cc = np.zeros(cant-1)
mean = np.zeros(cant-1)

for i in range(1,cant):
    cc[i-1] = i*0.1
    clf = svm.SVC(kernel='linear',C=cc[i-1])
    scores = cross_val_score(clf,X,Y,cv = 10)
    mean[i-1] = scores.mean()
    print("C = %f, Media: %0.16f , Std: %0.16f)" % (i*0.1, mean[i-1], scores.std()))
```

```python
print("max: "+str(max(mean)))
plt.plot(cc,mean,"r",color="blue")
```

* Por la izquierda tendencia a disminuir (0.1 que es un solo punto no representa la tendencia y esto se confirmar√° en el caso extremo c = 0.00001)
* Por la derecha tenemos 1, 2,3 ... que ya hemos calculado arriba y salia el mayor en c = 2,4 y 7 (Que igual se ve que la tendencia es a ir bajando conforme subimos el c)
* Al centro se tiene valores altos, el m√°ximo es en c = 0.8 y 1, siendo el escore en ese punto de 0.8815 que es parecido el mismo que cuando el valor cuando c fue 2.

## Variando los kernels y sus par√°metros

### Kernel Radial

```python
cs = [0.001, 0.01, 1, 10,50]
gammas =  [0.0001, 0.01, 1, 10, 50]

for i in cs:
    for j in gammas:
        clf = svm.SVC(kernel='rbf', C=i, gamma=j)   #0.0001
                                                        # c 10
        scores = cross_val_score(clf,X,Y,cv = 5)
        print("C = %0.4f, gamma: %0.4f, Media: %0.16f , Std: %0.16f)" % (i,j , scores.mean(), scores.std())) #0.8842023693605972
```

**Mejores Par√°metros**

C = 10

gamma = 0.0001

Media del score gracias a cross-validation = 0.8737861084063615



### Kernel Polinomial

**Nota:** Aqui hay que hacer un escalado a nuestros features, caso contrario nos tomar√° mucho tiempo ajustar nuestro modelo.

```python
from sklearn import preprocessing

X_new = preprocessing.scale(X)
```

```python
cs = [0.1, 1, 10, 20, 50]
dgs = [2,3,5,8,20]

for i in cs:
    for j in dgs:
        clf = svm.SVC(kernel='poly', C=i, degree=j, gamma='auto')  
        
        scores = cross_val_score(clf,X_new,Y,cv = 10)
        print("C = %0.4f, degree: %0.4f, Media: %0.16f , Std: %0.16f)" % (i,j , scores.mean(), scores.std()))
```

Notar que aqui cuando el degree es par, disminuye considerablemente el score. (c = 0.1, degree 8 -> score 0.7504790823211877)

**Mejores Par√°metros**

C = 10

degree = 3

Media del score gracias a cross-validation = 0.9117813765182188

## Gr√°ficando

Nos pide ahora el libro que acerca de nuestras evaluaciones de los par√°metros anteriores, gr√°fiquemos.

Esto para hacerlo ordenadamente se har√° con 3 casos:

* Lineal: C = 1
* Rbf: C = 10, gamma = 0.0001
* Poly: C = 10, degree = 3

Y en cada uno de estos como tenemos no solo 2 features asi que graficar features vs target no tiene sentido aqui por lo que se har√° cada feature con la salida mpg01.

###  Kernel: Lineal 
#### C:1

```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.gridspec as gridspec
```

```python
clf = svm.SVC(kernel='linear',C=1)

yy = df["mpg01"]
```

```python
gs = gridspec.GridSpec(3, 3)

columns = ['cylinders', 'displacement', 'horsepower', 'weight',
           'acceleration', 'year','origin_1', 'origin_2','origin_3']

dir_col = np.array(range(9))
dir_col = dir_col.reshape(3,3)

mpg_g = df.columns.to_list()[0]

fig = plt.figure(figsize=(12,10))

for i in range(3):
    for j in range(3):
        xx = df[[columns[dir_col[i,j]],mpg_g]]
        clf.fit(xx.values,yy.values)
        ax = plt.subplot(gs[i,j])
        fig = plot_decision_regions(X=xx.values, y=yy.values, clf=clf)
        plt.title(columns[dir_col[i,j]])
```
