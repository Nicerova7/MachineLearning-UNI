```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D 
```

```python
np.random.seed(17)
```

```python
def mean(a):
    return np.sum(a)/len(a)

def random_point(dim):
    return np.array([np.random.rand() for _ in range(dim)])

def distance(a,b):
    c = a-b
    return np.sqrt(np.dot(c,c))

def random_distances(dim,num_pairs):
    return [distance(random_point(dim), random_point(dim))
            for _ in range(num_pairs)]
```

```python
dimensions = range(1,101)

avg_distances = []
min_distances = []

for dim in dimensions:
    distances = random_distances(dim,10000) #10mil distancias
    avg_distances.append(mean(distances))
    min_distances.append(min(distances))
```

### 12.5

```python
plt.xlabel("N° dimension")
plt.ylabel("Distances")
plt.title("10mil Random Distances")
plt.plot(dimensions,avg_distances)
plt.plot(dimensions,min_distances)
plt.legend(["Average distance","Minimal distance"])
```

Podemos ver como en promedio (promedio de 10000 puntos) al aumentar la dimension de un par de puntos, aumenta la distancia entre ellos, aquí se compara con los valores mínimos de 10000 distancias y esto es para cada dimensión; Con esto se ve que la tendencia sigue hacia arriba es decir a mayor dimensión, mayor distancia.

### 12.6

```python
avg_per_min = np.array(min_distances)/np.array(avg_distances)
```

```python
plt.xlabel("N° dimension")
plt.ylabel("Distances")
plt.title("Minimal distances / Average distances ")
plt.plot(dimensions,avg_per_min)
```

En este gráfica tenemos el ratio entre la distancia minima y la distancia en promedio de los 10000 datos por cada dimensión, esto es bastante preocupante porque vemos como a medida que aumenta la dimensión el mínimo y el promedio se acercan a 1, indicando entonces la igualdad entre ellos lo que conlleva a que el poder predictivo del algoritmo del vecino más cercano disminuiría, ya que si el vécino más cercano es casi igual que cualquier vecino al rededor tenemos que no estaríamos manejando bien el más cercano sino que estaríamos agarrando cualquier punto al rededor.

Aquí se generó 50 puntos aleatorios entre [0,1] y de 1 sola dimensión: Podemos ver como estos puntos (representados por una linea para mejor visualización) estan agrupados muy cerca, algunos hasta parece que se superponen, lo que es traducido como: *están bien cerca*.

#### 12.8

```python
fifty_points_two_dimension = []

for i in range(50):
    fifty_points_two_dimension.append(random_point(2))

xs = []
ys = []
for i in fifty_points_two_dimension:
    xs.append(i[0])
    ys.append(i[1])
```

```python
plt.scatter(xs,ys)
```

Al generar 50 puntos de 2 dimensiones (cada dimensión entre [0,1]) y ponerlos en una gráfica observamos como ya no hay la misma cantidad de puntos superpuestos como en 1 dimensión, es decir: *Aumentamos una dimensión, y la distancia entre ellos se expandió un poco*.

#### 12.9

```python
fifty_points_three_dimension = []

for i in range(50):
    fifty_points_three_dimension.append(random_point(3))

xs,ys,zs = [],[],[]
for i in fifty_points_three_dimension:
    xs.append(i[0])
    ys.append(i[1])
    zs.append(i[2])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs,ys,zs)
plt.show()
```

Ya en 3 dimensiónes vemos como la tendencia de estos puntos se fueron expandiendo más: *A más dimensión, aumenta la distancia, pero si mantenemos el mismo numero de púntos* ya que si aumentamos la cantidad de puntos si veremos una imaǵen tán densa como cuando la dimensión fue 2 y si seguimos aumentando el número de puntos veremos como la densidad llegaría a ser tanto como en 1 dimensión, pero este *aumento* de puntos debe ser de manera exponencial ¿Porqué exponencial? Así se indica en el libro *Data Science par la par la partique: Fondamentaux avec Python.* Y se puede ver fácilamente al gráficarlo aumentando exponencial los puntos puntos al aumentar la dimensión
