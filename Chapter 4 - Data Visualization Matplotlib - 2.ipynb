### Diagrama de pastel

#### Personalizando el color

```python
%matplotlib inline
import matplotlib.pyplot as plt

labels = ["Chrome", "Internet Explorer","Firefox", "Edge","Safari","Sogou Explorer","Opera","Others"]

marketshare = [61.64, 11.98, 11.02, 4.23, 3.79, 1.63, 1.52, 4.19]

explode = (0,0,0.5,0,0.8,0,0,0)

colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']


plt.pie(marketshare,
        explode = explode,
        
        labels = labels,
        colors = colors,
        autopct="%.1f%%",
        
        shadow=True,
        startangle = 45)

plt.axis("equal")
plt.title("Web Browser Marketshare - 2018")
plt.show()
```

### Rotando el diagrama de pastel

Esto se hará gracias al parámetro startangle

```python
labels = ["Chrome", "Internet Explorer","Firefox", "Edge","Safari","Sogou Explorer","Opera","Others"]

marketshare = [61.64, 11.98, 11.02, 4.23, 3.79, 1.63, 1.52, 4.19]

explode = (0,0,0.5,0,0.8,0,0,0)

colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']


plt.pie(marketshare,
        explode = explode,
        
        labels = labels,
        colors = colors,
        autopct="%.1f%%",
        
        shadow=True,
        startangle = 0)

plt.axis("equal")
plt.title("Web Browser Marketshare - 2018")
plt.show()
```

### Mostrando leyenda

Aqui plt.pie retorna 3 valores:

* patches : una lista de instancias de matplotlib.patches.Wedge
* texts : una lista de instancias de la etiqueta de matplotlib.text.Text 
* autotexts : Una lista de instancias de Text para etiquetas numéricas, esto solo será retornado si el parámetro autopct is None.

```python
pie = plt.pie(marketshare,
              explode = explode,
              labels = labels,
              colors = colors,
              autopct="%.1f%%",
              shadow=True,
              startangle=45)

plt.axis("equal")
# turns off the axis lines and labels
plt.title("Web Browser Marketshare - 2018")
plt.legend(pie[0], labels, loc="best") # leyenda
plt.show()
```

El orden en la leyenda será configurado por el parámetro loc, en este caso tenemos al mejor primero.

### Guardando el chart 

Como ya se vió, podemos manipular distintos gráficos a nuestra conveniencia por lo que ahora nos falta guardar estos gráficos en nuestra computadora.

```python
labels = ["Chrome", "Internet Explorer","Firefox", "Edge","Safari","Sogou Explorer","Opera","Others"]

marketshare = [61.64, 11.98, 11.02, 4.23, 3.79, 1.63, 1.52, 4.19]

explode = (0,0,0.5,0,0.8,0,0,0)

colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']


plt.pie(marketshare,
        explode = explode,
        
        labels = labels,
        colors = colors,
        autopct="%.1f%%",
        
        shadow=True,
        startangle = 0)


plt.axis("equal") 
plt.title("Web Browser Marketshare - 2018") 
plt.savefig("Webbrowsers.png", bbox_inches="tight") # GUARDANDO y tight para quitar espacio extra.
plt.show()
```

### Graficando Scatter plots

Un scatter plot es un gráfico bidimensional que usa puntos u otras formas para representar valores, estos  gráficos son usado comúnmente para ver como una variable afecta a otra.

```python
plt.plot([1,2,3,4],
         [1,8,27,64],
         'bo')
plt.axis([0, 4.5, 0, 70]) # axis([xi,xf, yi, yf])
plt.show()
```

### Combinando plots

```python
import numpy as np
```

```python
a = np.arange(1,4.5,0.1)

plt.plot(a, a**2, 'y^',  #1
         a, a**3, 'bo',  #2
         a, a**4, 'r--',) #3
plt.axis([0, 4.5, 0, 70])
plt.show()
```

### Subplots

```python
a = np.arange(1,5,0.1)

plt.subplot(121) # 1 fila 2 columnas, 1er gráfico
plt.plot([1,2,3,4,5],
         [1,8,27,64,125],
         'y^') # tipo de gráfico triángulo

plt.subplot(122)  # 1 fila , 2 columnas , 2do gráfico
plt.plot(a, a**2, 'y^',
         a, a**3, 'bo',
         a, a**4, 'r--',)

plt.axis([0, 4.5, 0, 70])
plt.show()
```
