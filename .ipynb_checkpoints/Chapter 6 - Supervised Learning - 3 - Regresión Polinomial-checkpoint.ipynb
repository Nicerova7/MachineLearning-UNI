## Regresión Polinomial

Hemos visto que cuando usamos una regresión múltiple el valor R² no salía tan exacta o valores cercanos a uno, esto es porque la regresión lineal no siempre es la mejor solución para capturar la relación entre las variables independientes con la variable dependiente, por lo que en algunos casos usar una curva es la mejor solución.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('polynomial.csv')
plt.scatter(df.x,df.y)
```

Realizemos un modelo con una regresión lineal

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()

x = df.x[0:6, np.newaxis] # convertimos a array 2D , es decir lo convertimos a una columna
y = df.y[0:6, np.newaxis] # convertimos a array 2D

model.fit(x,y)
```

```python
y_pred = model.predict(x)

plt.scatter(x, y, s=10, color='b')

# dibujamos los puntos predecidos
plt.plot(x, y_pred, color='r') 
plt.show() 

# Calculamos el R-cuadrado
print('R-cuadrado para el set de entrenamiento es: %.4f' %model.score(x,y))
```

Como podemos observar el valor R-cuadrado es bastante alto pero queremos ver si cambiando una regresión lineal por una polinomial este estadístico R-cuadrado mejorará.

### Fórmula para la Regresión Polinomial

![](formulapoly.png)

   * Y: Variable dependiente
   * $\beta$<sub>0</sub>: intercepto
   * $\beta$<sub>1</sub>, $\beta$<sub>2</sub>, $\beta$<sub>3</sub>, ... ,$\beta$<sub>n</sub>: Coeficientes para las n-ésimas variables independientes.
   
La idea en de éste método es el mismo: Buscar los coeficientes que mejor ajusten los datos.

### Regresión Polinomial en Scikit-Learn

En la librería Scikit-Learn tenemos un número de clases y funciones para resolver regresiones polinomiales.

```python
from sklearn.preprocessing import PolynomialFeatures

# Especificamos el grado de nuestra formula:
degree = 2
polynomial_features = PolynomialFeatures(degree=degree)

x_poly = polynomial_features.fit_transform(x)
print(x_poly)
```

   * La primera columna siempre contiene 1
   * La segunda columna es el valor de x (los valores de x que hemos cargado del archivo .csv)
   * La tercera columna es el valor de x² (elevamos al cuadrado los valores anteriores)
   
Usaremos esta matriz con la clase LinearRegression() para entrenar el modelo:

```python
model = LinearRegression()
model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)

# Graficamos
plt.scatter(x, y, s=10)
plt.plot(x,y_poly_pred)
plt.show()
```

```python
# Imprimimos algunos valores importantes
print(model.coef_)
print(model.intercept_)
```

Los valores y dado un x son calculados por: Y = -0.87153912 + 1.98293207*x + 0.17239897*x²

```python
print('R-cuadrado por set de entrenamiento: %.4f' %model.score(x_poly,y))
```

¿ Puede mejorar esto, aún más? Intentemos con un polinomio de grado 5.

```python
degree = 5
polynomial_features = PolynomialFeatures(degree=degree)

x_poly = polynomial_features.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)

y_poly_pred = model.predict(x_poly)

plt.scatter(x, y, s=10)
plt.plot(x, y_poly_pred)
plt.show()
```

```python
model.score(x_poly,y) # Nos resulto un puntaje perfecto
```

Aunque el estadístico R-cuadrado es 1.0 es poco probable que con nuevos datos resulte con el mismo R-cuadrado, a este error se llama **sobreajustamiento** o en inglés **overfitting**.

### Entendiendo Bias y Variance

La incapacidad de los algoritmos de *aprendizaje automático* de capturar la verdadera relación entre las variables y el resultado es conocido como Bias:
   * Bias alto: La curva no pasa por todos los puntos ![](highbias.png)
   * Bias bajo: La curva pasa por todos los puntos ![](lowbias.png)
   
Pero que sucede si seguimos agregando data del dataset (más valores que no hemos tomado en cuenta para entrenar)

   * Variance bajo: Los puntos nuevos presentan un error demasiado alto (muy alejados de la curva) ![](lowvariance.png)
   * Variance alto: Los puntos nuevos presentan un error pequeño (siguen apegados a la curva) ![](highvariance.png)
   
La RSS o suma residual al cuadrado es muy bajo cuando se tiene una variance bajo, mientras que es totalmente opuesto con una variance alta.

En el *aprendizaje automático*, el ajuste entre distintos conjuntos de datos es conocido como **variance**, vemos como una curva tiene una alta **variance** por lo que no podremos realmente tener la certeza de cuan bien predecirá nuestro modelo cuando en un futuro agregemos más datasets, por otro lado una linea tiene una **variance** baja por lo que el RSS en estos casos son menores para diferentes datasets.

**NOTA** : Si bien existe el overfitting que es cuando tenemos la curva pasando por todos los puntos, también existe el underfitting que es cuando tenemos la curva(linea) que no pasa por ningún punto.

### Lo ideal

Podemos entonces resumir diciendo que necesitamos un modelo con:

   * Alto bias.
   * Baja variance.
   ![](idealcurva.png)
   
Para llegar a esto podremos realizar algunas técnicas como:
   * Regularization: Técnica que automáticamente penaliza las extras variables independientes usados en el modelo.
   * Bagging: Tipo de proceso de *aprendizaje automático* que usa un aprendizaje en conjunto que evoluciona los modelos. Se usa un subset de la data en cada débil *learner* para que luego todos los *learners* débiles se combinen para crear un fuerte *learner* que tiene exactas predicciones.
   * Boosting: Similar al bagging, excepto que se usa todo el dataset para entrenar cada *learner*, pero los datapoints que fueron mál clasificados tienen mayor peso para que los futuros *learners* puedan tener mayor enfoque en estos datapoints durante el entrenamiento.

### Usando Regresión múltiple polinomial en el dataset Boston

```python
from sklearn.datasets import load_boston

dataset = load_boston()

df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df['MEDV'] = dataset.target

# de todos las columnas nos quedamos tan solo con las que más correlación se tiene con la variable dependiente.
x = pd.DataFrame(np.c_[df['LSTAT'], df['RM']], columns= ['LSTAT','RM'])
Y = df['MEDV']

# extramoes el 70% de los datos para entrenar y el 30% para testear.
from sklearn.model_selection import train_test_split
x_train, x_test, Y_train, Y_test = train_test_split(x, Y, test_size = 0.3, random_state = 5)

degree = 2
polynomial_features = PolynomialFeatures(degree=degree)
x_train_poly = polynomial_features.fit_transform(x_train)
```

Como ahora es polinomial pero tenemos dos variables independientes (x1 y x2) la formula varia un poco
![](polymulti.png)

```python
print(polynomial_features.get_feature_names(['x','y']))
```

```python
model = LinearRegression() # entrenmos nuestro modelo con estos últimos valores dados...
model.fit(x_train_poly, Y_train)
```

```python
# evaluamos nuestro modelo con los datos para testear...
x_test_poly = polynomial_features.fit_transform(x_test)
print('R-cuadrado: %.4f' %model.score(x_test_poly, Y_test))
```

```python
# Ahora imprimimos el intercepto y los coeficientes:
print(model.intercept_)
model.coef_
```

El modelo resultaría alfinal:

Y = 26.9334305238 + 1.47424550e+00*x<sub>1</sub> + (-6.70204730e+00)*x<sub>2</sub> + 7.93570743e-04*x²<sub>1</sub> + (-3.66578385e-01)*x<sub>1</sub>x<sub>2</sub> + 1.17188007e+00*x²<sub>2</sub>

### Gráficando el hiperplano en 3D

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (18,15))
ax = fig.add_subplot(111, projection = '3d')

ax.scatter(x['LSTAT'],
           x['RM'],
           Y,
           c='b')

ax.set_xlabel("LSTAT")
ax.set_ylabel("RM")
ax.set_zlabel("MEDV")

# Creamos el meshgrid para poder rotar en 3D nuestra imagen (aunque eso desde terminal, no aquí en jupyter)
x_surf = np.arange(0, 40, 1) # LSTAT
y_surf = np.arange(0, 10, 1) # RM
x_surf, y_surf = np.meshgrid(x_surf, y_surf)

z = lambda x,y: (model.intercept_ +
                 model.coef_[0]*x +
                 model.coef_[1]*y +
                 model.coef_[2]*x*x + 
                 model.coef_[3]*x*y + 
                 model.coef_[4]*y*y)

ax.plot_surface(x_surf, y_surf, z(x_surf,y_surf),
                rstride = 1,
                cstride = 1,
                color = 'None',
                alpha = 0.4)

plt.show()
```

## Resumen

   * Regresión Múltiple: Regresión Lineal entre dos o más variables independientes y una variable dependiente.
   * Regresión Polinomial: Regresión entre una variable dependiente y una variable dependiente usando polinomio de grado n-ésimo.
   * Regresión Múltiple Polinomial: Modelado de la relación entre dos o más variables independientes con una variable dependiente usando un polinomio de grado n-ésimo.
   
También se vió como realizar el gráfico de un hiperplano en 3D mostrando la relación entre variables independientes y su salida respectiva.
