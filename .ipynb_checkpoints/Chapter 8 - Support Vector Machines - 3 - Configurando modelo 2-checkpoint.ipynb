### Radial Basis Function (RBF) Kernel

Además del kernel lineal ('linear') que hemos visto existen otros como:

* Radial Basis function (RBF) también conocido como Gaussian Kernel
* Polynomial

#### RBF:
Da valor a cada punto en función de su distancia desde el origen o un centro fijo, comúnmente en un espacio euclidiano. 

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import svm,datasets
```

```python
iris = datasets.load_iris()
```

```python
C = 1 

X = iris.data[:, :2]
y = iris.target

clf = svm.SVC(kernel='rbf', gamma='auto', C=C).fit(X, y)
title = 'SVC con kernel RBF '
```

### Gamma 

En la última celda de nuestro notebook hemos introducido un nuevo parámetro: gamma.
Gamma define hasta qué punto la influencia de un solo ejemplo de entrenamiento alcanza.

Un valor Gamma bajo indica que cada punto tiene un alcance lejano.

![](lowgamma.png)

Un Gamma alto significa que los puntos más cercanos al límite de decisión tienen un alcance cercano. Cuanto mayor sea el valor de Gamma, más intentará ajustarse exactamente al conjunto de datos de entrenamiento, lo que resulta en un **sobreajuste**.

![](highgamma.png)



![](Cgamma.png)
Para resumir, C controla la suavidad del límite y Gamma determina si los puntos están sobreajustados.

### Kernel Polinomial

Un kernel polinomial de grado 1 es similar al del kernel lineal. Los kernel polinomiales de mayor grado ofrecen un límite de decisión más flexible.

```python
C = 1 # SVM parámetro de regularización

clf = svm.SVC(kernel='poly', degree=4, C=C, gamma='auto').fit(X, y)
title = 'SVC con kernel polinomial (grado 4)'
```

```python
# máximo y minimo del primer feature
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1

# máximo y minimo del segundo feature
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

# tamaño de paso de la malla
h = (x_max / x_min)/100

# haciendo predicciones para cada punto en xx,yy
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])


# dibujando el resultado usando un color plot
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Accent, alpha=0.8)

# dibujando los puntos de entrenamiento 
colors = ['red', 'green', 'blue']
for color, i, target in zip(colors, [0, 1, 2], iris.target_names):
    plt.scatter(X[y==i, 0], X[y==i, 1], color=color, label=target)

plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title(title)
plt.legend(loc='best', shadow=False, scatterpoints=1)
```

### SVM Para problemas de la vida real

```python
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn import svm
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(font_scale=1.2)
```

```python
data = pd.read_csv('house_sizes_prices_svm.csv')

sns.lmplot('size', 'price',data=data,hue='sold',palette='Set2',fit_reg=False,scatter_kws={"s": 50});
```

```python
X = data[['size','price']].values
y = np.where(data['sold']=='y', 1, 0) # 1 : Y  ; 0 para N

model = svm.SVC(kernel='linear').fit(X, y)
```

```python
# primer feature
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1

# segundo feature
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1


# tamaño de paso para malla
h = (x_max/x_min)/ 20

#  predicciones para cada punto en xx,yy
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])

colors = ['red', 'blue']
for color, i, target in zip(colors, [0, 1], y):
    plt.scatter(X[y==i, 0], X[y==i, 1], color=color, label=target)
    
    
# dibuja el resultado unsado un color de:
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Blues, alpha=0.3)


plt.xlabel('Tamaño de casa')
plt.ylabel('Asking price (1000s)')
plt.title("Size of Houses and Their Asking Prices")
```

```python
def will_it_sell(size, price):
    if(model.predict([[size, price]]))==0:
        print('No venta!')
    else:
        print('Venta, cachinnnn!')
```

```python
will_it_sell(2500, 400) 
will_it_sell(2500, 200)
```
