## Regresión Múltiple

Ahora veremos como entrenar un modelo usando varias variables independientes y una variable dependiente, esto es conocido como regresión múltiple. En pocas palabras se usará dos o más variables independientes para predecir una variable dependiente.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_boston

dataset = load_boston() 
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df['MEDV'] = dataset.target

df.head()
```

```python
# Relación entre LSTAT y MEDV (variable dependiente)

plt.scatter(df['LSTAT'], df['MEDV'], marker = 'o')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
```

```python
# Relación entre RM y MEDV (variable dependiente)
plt.scatter(df['RM'], df['MEDV'], marker='o')
plt.xlabel('RM')
plt.ylabel('MEDV')
```

```python
# Para una mejor visualización veamos estas dos variables independientes (RM y LSTAT) con MEDV (variable dependiente)
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(18,15))
ax = fig.add_subplot(111, projection = '3d')

ax.scatter(df['LSTAT'],
           df['RM'],
           df['MEDV'],
           c = 'b')

ax.set_xlabel("LSTAT")
ax.set_ylabel("RM")
ax.set_zlabel("MEDV")
plt.show()
```

Ahora que hemos visualizado nuestros datos, podemos entrenar un modelo en base a estos datos.

```python
x = pd.DataFrame(np.c_[df['LSTAT'],df['RM']], columns=['LSTAT','RM']) # un dataframe con LSTAT y RM (np.c_ es concatenar)
Y = df['MEDV'] # otro dataframe con la variable dependiente.
```

```python
# De todos los datos 70% o 0.7 serán usado para entrenarlos y 30% o 0.3 usados para test o datos de prueba (si nuestro
# modelo esta bien o no)
from sklearn.model_selection import train_test_split
x_train, x_test, Y_train, Y_test = train_test_split(x, Y, test_size = 0.3, random_state = 5)

# Random_state será usado como una semilla para saber que datos fueron sacados aleatoriamente para ser el 70% de train.
```

```python
print(x_train.shape) # esto nos dice que la cantidad de valores para ser usados en el entrenamiento son 354
print(Y_train.shape)
```

```python
print(x_test.shape) # esto nos dice que la cantidad de valores pasa ser usados en el test o prueba son 152
print(Y_test.shape) 
```

Ahora que hemos dividido nuestro los valores de nuestro dataframe procedemos a entrenar con scikit-learn

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, Y_train)
```

```python
# Una vez que hemos entrenado nuestro modelo , procedemos a probar con nuestros valores test 

price_pred = model.predict(x_test) 
```

Usaremos el estadístico **R²** para ver que tan bien esta nuestro modelo.

```python
print('R-squared: %.4f' % model.score(x_test,Y_test))
```

Este valor 0.6162 nos indica que tan bien esta predecido nuestro modelo como es un valor mayor que 0.5 podemos decir que esta relativamente bien... recordar que **R-cuadrado, se define como la proporción de la varianza total de la variable explicada por la regresión**.

```python
# Tambien podemos calcular el error cuadrático medio que es el promedio de los errores al cuadrado 
# (el promedio de la diferencia entre el estimador y el estimado pero al cuadrado)

from sklearn.metrics import mean_squared_error

mse = mean_squared_error(Y_test, price_pred)

mse2 = np.sum((Y_test - price_pred)**2) / len(Y_test) # modo manual
print(mse)
print(mse2)
```

```python
plt.scatter(Y_test, price_pred)
plt.xlabel("Precio Actual")
plt.ylabel("Precio predecido")
plt.title("Precio actual vs Precio predecido")
```

### Obteniendo el intercepto y los coeficientes

![](multiplereg.png)

Donde:
   * Y: variable dependiente
   * $\beta$<sub>0</sub>: El intercepto
   * $\beta$<sub>1</sub> y $\beta$<sub>2</sub>: Coeficientes de las variables independientes que se toman del dataframe. 

```python
print(model.intercept_)
print(model.coef_)
```

```python
# podemos usar un valor para ver cual valor predice nuestro modelo :
print(model.predict([[30,5]]))
```

```python
# podemos hacerlo "manualmente" con los coeficientes dadas por el modelo

# y = b_0               +  b_1 * x1      +  b_2 * x2
y = 0.38437936780346504 + -0.65957972*30 +  4.83197581*5
y
```

### Graficando en 3D un hiperplano

Veamos como se ve la predicción en un hiperplano 3D

```python
fig = plt.figure(figsize=(18,15))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x['LSTAT'],
           x['RM'],
           Y,
           c = 'b')

ax.set_xlabel("LSTAT")
ax.set_ylabel("RM")
ax.set_zlabel("MEDV")

# creamos un meshgrid para graficarlo en 3D y poder rotarlo (aquí no se podra realizar esto porque jupyter notebook es
# estático) 
# si queremos verlo en movimiento podemos realizarlo en el terminal y no en jupyter notebook

x_surf = np.arange(0, 40, 1)
y_surf = np.arange(0, 10, 1)
x_surf, y_surf = np.meshgrid(x_surf, y_surf)

z = lambda x,y: (model.intercept_ + model.coef_[0]*x + model.coef_[1]*y)  # creamos la prediccion en base a los coef.

ax.plot_surface(x_surf, y_surf, z(x_surf,y_surf),
                rstride = 1,
                cstride = 1,
                color = 'None',
                alpha = 0.4)

plt.show()
```
