## Kernel Trick
Consideremos el siguiente conjunto: 

```python
%matplotlib inline
import numpy as np 
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from sklearn.datasets.samples_generator import make_circles
X,c = make_circles(90, 
                   factor=0.2, # factor de escala entre los dos circulos
                   noise=0.1, 
                   random_state=5 # semilla
                  ) 

plt.scatter(X[:,0],X[:,1], c=c, s=80, cmap='seismic')
plt.show()
```

Claramente los puntos rojos no pueden ser separados linealmente de los puntos azules. Es decir, no es posible dibujar una linea recta que separe a los conjuntos de puntos. 

Una forma de poder emplear SVM en este dataset es imaginar que sacudimos el plano en el que se encuentra de tal manera que los puntos rojos se esten separados de los azules en el eje z.

![Shake](shake.gif "Shake it")

Ahora, ya es posible dibujar un plano (en lugar de una linea ya que ahora trabajamos en 3d) para poder clasificar nuestros puntos. A esta t√©cnica que transforma los datos en un espacio de dimensi√≥n superior se le conoce como **Kernel Trick**. Como ya hemos visto, con esta t√©cnica se consigue un claro margen de divisi√≥n entre las clases de datos. 

```python
# ==== Agregando la tercera dimension de forma manual ====
fig = plt.figure(figsize=(18,15))
ax = fig.add_subplot(111, projection='3d')
z = X[:,0]**2 + X[:,1]**2 # z = x^2 + y^2
ax.scatter(X[:, 0], X[:, 1], z, c=c, s=80, cmap='seismic')
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# ==== Entrenando el modelo usando la 3era dimension ====

# Combinando 'X' y 'z' en una matriz
features = np.concatenate((X,z.reshape(-1,1)), axis=1)

# Entrenando usando SVM lineal
from sklearn import svm
clf = svm.SVC(kernel = 'linear')
clf.fit(features, c)

# Graficando el hiperplano
x3 = lambda x,y: (-clf.intercept_[0] - clf.coef_[0][0] * x-clf.coef_[0][1] * y) / clf.coef_[0][2]

tmp = np.linspace(-1.5,1.5,100)
x,y = np.meshgrid(tmp,tmp) # Retorna las combinaciones de tmp consigo mismo
ax.plot_surface(x, y, x3(x,y),color='skyblue', alpha = 0.6)
plt.show()
```

La formula del hiperplano es ahora:

$$g(x) = \overrightarrow{w_0}x_1 + \overrightarrow{w_1}x_2 + \overrightarrow{w_2}x_3 + b$$

donde:

- $\overrightarrow{w_0}$ toma el valor de `clf.coef_[0][0]`.
- $\overrightarrow{w_1}$ toma el valor de `clf.coef_[0][1]`.
- $\overrightarrow{w_2}$ toma el valor de `clf.coef_[0][2]`.
- $b$ toma el valor de `clf.intercept_[0]`.

Para graficar el hiperplano es necesario encontrar el valor de $x_3$:

$x_3 = \frac{-\overrightarrow{w_0}x_1 - \overrightarrow{w_1}x_2 - b}{\overrightarrow{w_2}}$

## Kernels
A veces, no todos los puntos se pueden separar linealmente, ni se pueden separar utilizando ***Kernel Trick*** como en el √∫ltimo ejemplo. Para este tipo de datos, se deben *curvar* las l√≠neas para separarlas. 

En ML, los ***kernels*** son funciones que transforman los datos de espacios no lineales a lineales. 

![kernel](kernel.PNG)

Suponga que deseamos aplicar una transformaci√≥n polinomial de 2do grado a un conjunto de entrenamiento bidimensional, luego entrenamos un clasificador lineal SVM en el conjunto de entrenamiento transformado. La funci√≥n de mapeo polinomial de 2do grado $\phi$ que deseamos aplicar es:

![phi](phi.PNG "phi")

Ahora, veamos qu√© pasa si aplicamos este mapeo a los vectores $a$ y $b$ y seguidamente calculamos su producto punto. 

![kernel punto](kernel_punto.PNG "kernel punto")

Si aplicamos la transformaci√≥n $\phi$ a todos los ejemplos de entrenamiento, encontraremos la expresi√≥n anterior en la funci√≥n que deseamos minimizar (**forma dual del objetivo del SVM lineal**). As√≠ no estar√≠amos transformando cada ejemplo de entrenamiento sino solo estar√≠amos reeemplazando el producto punto por su valor elevado al cuadrado.

La funci√≥n

$$K(a,b) = (a^Tb)^2$$

es conocida como el kernel polinomial de grado 2.

As√≠, tenemos una definici√≥n m√°s precisa del kernel. En ML, el kernel es una funci√≥n capaz de calcular el producto interno $\phi(a)^T\phi(b)$ bas√°ndose en √∫nicamente los vectores originales $a$ y $b$, sin tener que calcular la transformaci√≥n $\phi$.

Los kernels m√°s usados son:

- Lineal 

$$K(a,b) = a^Tb$$

- Polinomial

$$K(a,b) = (\gamma a^Tb+r)^d$$

- Gausiano

$$K(a,b) = exp(-\gamma \|a-b\|^2)$$

```python
%matplotlib inline
import pandas as pd
import numpy as np

from sklearn import svm, datasets
import matplotlib.pyplot as plt
iris = datasets.load_iris()

# ==== Definimos la data ====
X = iris.data[:, :2] # Tomamos solo los dos primeros features
                     # ('sepal length' y 'sepal width')
y = iris.target # 'setosa'(0) o 'versicolor'(1) o 'virginica'(2)

# ==== Graficando los puntos de entrenamiento ====
colors = ['red', 'green', 'blue']

for color, i, target in zip(colors, [0, 1, 2], iris.target_names):
    plt.scatter(X[y==i, 0], X[y==i, 1], color=color, label=target)
    
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('Scatter plot of Sepal width against Sepal length')
plt.show()
```

**Par√°metro de regularizaci√≥n C**

- Un **alto** valor de C har√° que el algoritmo se centre m√°s en **clasificar correctamente los puntos.** La desventaja de esto es que puede resultar en un **margen m√°s estrecho**. 


- Un **bajo** valor de C har√° que el algoritmo apunte al **margen m√°s amplio posible**, pero puede clasificaalgunos puntos se **clasifiquen incorrectamente**.

## Kernel Lineal

```python
# ==== Usando SVM con el kernel lineal ====

C = 1 # parametro de regularizacion
clf = svm.SVC(kernel='linear', C=C).fit(X, y)
title = 'SVC with linear kernel'

# Minimo y maximo
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1 # Sepal length
print('x_min={0}, x_max={1}'.format(x_min,x_max))
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1 # Sepal width
print('y_min={0}, y_max={1}'.format(y_min,y_max))

# Paso en el mesh
h = (x_max / x_min)/100

# ==== Clasificando los puntos a partir de las predicciones para cada punto en xx,yy ====

# Combinaciones de las secuencias de valores para cada eje
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), # eje x
                     np.arange(y_min, y_max, h)  # eje y
                    ) 

# Toma  valores de 0 ('setosa'), 1 ('versicolor') o 2('virginica')
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]) 

# ==== Dibujando contornos rellenos para =identificar los limites de las clases ===
Z = Z.reshape(xx.shape) # Z debe tener las mismas dimensiones que xx e yy
plt.contourf(xx, yy, Z, 
             cmap=plt.cm.Accent, 
             alpha=0.8 # Transparencia
            )

# ==== Graficando los puntos de entrenamiento ====
colors = ['red', 'green', 'blue']

for color, i, target in zip(colors, [0, 1, 2], iris.target_names):
    plt.scatter(X[y==i, 0], X[y==i, 1], color=color, label=target)
    
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title(title)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.show()
```

```python
# Realizando predicciones sobre todos los puntos de entrenamiento

predictions = clf.predict(X)
print(iris.target_names)
print(np.unique(predictions, return_counts=True)[1])
```

**Par√°metro Gamma**
- Un valor **bajo** de Gamma permite que **cada punto** tenga el mismo alcance.

- Un valor **alto** de Gamma se enfoca m√°s en **puntos cercanos al l√≠mite**.

## Kernel Polinomial

Un kernel polinomial de grado 1 es similar al kernel lineal. Los kernel polinomiales de mayor grado ofrecen un decision boundary m√°s flexible.

```python
# ==== Usando SVM con el kernel lineal ====
def kernelPoly(degree):
    C = 1 # parametro de regularizacion
    clf = svm.SVC(kernel='poly', degree=degree, C=C, gamma='auto').fit(X, y)
    title = 'SVC with polynomial (degree '+str(degree)+') kernel'

    # Minimo y maximo
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1 # Sepal length
    print('x_min={0}, x_max={1}'.format(x_min,x_max))
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1 # Sepal width
    print('y_min={0}, y_max={1}'.format(y_min,y_max))

    # Paso en el mesh
    h = (x_max / x_min)/100

    # ==== Clasificando los puntos a partir de las predicciones para cada punto en xx,yy ====

    # Combinaciones de las secuencias de valores para cada eje
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), # eje x
                         np.arange(y_min, y_max, h)  # eje y
                        ) 

    # Toma  valores de 0 ('setosa'), 1 ('versicolor') o 2('virginica')
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]) 

    # ==== Dibujando contornos rellenos para =identificar los limites de las clases ===
    Z = Z.reshape(xx.shape) # Z debe tener las mismas dimensiones que xx e yy
    plt.contourf(xx, yy, Z, 
                 cmap=plt.cm.Accent, 
                 alpha=0.8 # Transparencia
                )

    # ==== Graficando los puntos de entrenamiento ====
    colors = ['red', 'green', 'blue']

    for color, i, target in zip(colors, [0, 1, 2], iris.target_names):
        plt.scatter(X[y==i, 0], X[y==i, 1], color=color, label=target)

    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.title(title)
    plt.legend(loc='best', shadow=False, scatterpoints=1)
    plt.show()
```

```python
kernelPoly(1) # Kernel Lineal
```

```python
kernelPoly(2)
```

```python
kernelPoly(3)
```

```python
kernelPoly(4)
```
