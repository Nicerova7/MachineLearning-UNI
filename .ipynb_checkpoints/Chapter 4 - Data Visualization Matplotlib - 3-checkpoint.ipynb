## Data Visualization Matplotlib 3

### Usando Seaborn

Como se vió podemos usar plot para generar gráficos pero configurar los parámetros es a veces un poco complicado,
más cuando tratamos con grandes datos con distintas relaciones.

Seaborn es un librería de gráficos complementarios del paquete matplotlib.
La fuerza de Seaborn reside en su capacidad para crea gráficos estadísticos en Python, y está estrechamente integrado con los Pandas estructura de datos Seaborn proporciona abstracciones de alto nivel para permitirle crear visualizaciones complejas para sus datos fácilmente.

En conclusión: menos configuraciónes e imágenes más sofisticadas.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv('drivinglicense.csv')
```

```python
g = sns.catplot(x="gender", y="license", col="group",data=data, kind="bar", ci=None, aspect=1.0)

# Configurando los labels
g.set_axis_labels("", "Proportion with Driving license")
g.set_xticklabels(["Men", "Women"])
g.set_titles("{col_var} {col_name}")

plt.show()
```

Seaborn tiene distintos data sets para visualizarlos pondremos:

```python
sns.get_dataset_names()
```

```python
titanic = sns.load_dataset("titanic")

g = sns.catplot(x = "who", y = "survived", col ="class", data = titanic, kind = "bar", ci = None, aspect = 1)

g.set_axis_labels("", "Survival Rate")
g.set_xticklabels(["Men", "Women", "Children"])
g.set_titles("{col_name} {col_var}")

plt.show()
```

Se observó una data basado en clases y como se nota tanto como mujeres como niños de todas las clases se salvaron en mayor cantidad.

### Mostrando Lmplots

Otro dataset popular en seaborn es Lmplot que es un diagrama de dispersión, y un dataset de este tipo es iris que contiene información de flores.

```python
iris = sns.load_dataset("iris")

sns.lmplot('petal_width', 'petal_length', data=iris,
hue='species', palette='Set1',
fit_reg=False, scatter_kws={"s": 70})

ax = plt.gca()
ax.set_title("Grafico usando dataset iris") # no se puede poner tilde

plt.show()
```

### Visualización de Swarmplots

Un swarmplot es un diagrama de dispersión categórico con puntos que no se superponen, utilizado comúnmente para descubrir distribución de datos.

```python
sns.set_style("whitegrid")
```

```python
data = pd.read_csv('salary.csv')

sns.swarmplot(x="gender", y="salary", data=data)

ax = plt.gca()
ax.set_title("Distribucion de salario")

plt.show()
```

### Finalizando 

En este capítulo vimos distintos tipos de gráficos para distintos usos principalmente descubrir relaciones entre datos o patrones en un dataset.
